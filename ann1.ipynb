```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10,10)
plt.plot(x,binaryStep(x))
plt.axis('tight')
plt.title("Activation function : binary Step")
plt.show()
```

```python
def linear(x):
    return x
```

```python
plt.plot(x,linear(x))
plt.axis('tight')
plt.title("Activation function : linear")
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title("Activation function : sigmoid")
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title("Activation function : tanh")
plt.show()
```

```python
def Relu(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
```

```python
x = np.linspace(-10,10)
plt.plot(x,Relu(x))
plt.axis('tight')
plt.title("Relu")
plt.show()
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis = 0)
```

```python
plt.plot(x,softmax(x))
plt.axis('tight')
plt.title("softmax")
plt.show()
```

```python
def softplus(x):
    return np.log(1+np.exp(x))
```

```python
plt.plot(x,softplus(x))
plt.axis('tight')
plt.title("softplus")
plt.show()
```
